```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#reading the CSV and getting the head.
restaurant_df = pd.read_csv('Resources/DOHMH_New_York_City_Restaurant_Inspection_Results.csv')
restaurant_df.head()
```

```python
restaurant_df.info()
```

```python
## Drop unecessary columns
to_drop = ['PHONE', 'BUILDING','Community Board', 'Council District','Census Tract', 'BIN', 'BBL', 'NTA', 'CRITICAL FLAG', 'SCORE', 'GRADE DATE','RECORD DATE', 'INSPECTION TYPE']
new_restaurant_df = restaurant_df.drop(to_drop, axis=1)
new_restaurant_df.head()
```

```python
# Sum of missing values in each column
new_restaurant_df.isnull().sum()
```

```python
# Count the amount of rows that have duplication
new_restaurant_df.duplicated().sum()
```

```python
#Show the rows that have duplication
new_restaurant_df.loc[new_restaurant_df.duplicated(), :]
```

```python
# Drop null rows
final_restaurant_df = new_restaurant_df.dropna()
final_restaurant_df
```

```python
# After removing unecessary columns and nulll rows, the dataset is even.
final_restaurant_df.info()
```

```python
# Final df has no duplications
final_restaurant_df.duplicated().sum()
```

```python
# Draft final dataframe with the proposed columns dropped, removed NAs and reformatted the ZIPCODE column to int32
final_restaurant_df.ZIPCODE = final_restaurant_df.ZIPCODE.astype('int')
final_restaurant_df.head()
```

```python
#Final dataframe for restaurant dataset, removing remaining duplicates.
restaurant_final_table = final_restaurant_df.sort_values(by=['DBA']).drop_duplicates(subset=['DBA','ZIPCODE'])
restaurant_final_table.head()
```

```python
# First query, top zipcodes with the most rats
pd.read_sql_query('select * from rats order by rat_findings desc', con= engine).head()
```

```python
# Second query, Restaurants in zipcode 10027 and their violations and rat findings
pd.read_sql_query('''select r.camis, r.dba, r.boro, r.zipcode, r.grade, v.violation_code, v.violation_description, t.rat_findings
from restaurants r 
join inspections i on r.camis = i.camis
join violations v on i.violation_code = v.violation_code
join rats t on r.zipcode = t.zipcode
where r.zipcode = '10027'
''', con= engine).head()
```

```python
# Third query - Restaurants with a grade of P
pd.read_sql_query('''select r.camis, r.dba, r.boro, r.zipcode, r.cuisine, v.violation_code, v.violation_description, i.inspection_date, t.rat_findings
from restaurants r 
join inspections i on r.camis = i.camis
join violations v on i.violation_code = v.violation_code
join rats t on r.zipcode = t.zipcode
where r.grade = 'P'
order by r.grade desc
''', con=engine).head()
```

```python
# Fourth query - Restaurants with a grade of A
pd.read_sql_query('''select r.camis, r.dba, r.boro, r.zipcode, r.cuisine, v.violation_code, v.violation_description, i.inspection_date, t.rat_findings
from restaurants r 
join inspections i on r.camis = i.camis
join violations v on i.violation_code = v.violation_code
join rats t on r.zipcode = t.zipcode
where r.grade = 'A'
order by r.grade desc
''', con=engine).head()
```
