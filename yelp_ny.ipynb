```python
business_data=pd.read_json('resources/business.json', lines=True)
business_data
```

```python
yelp_df=business_data.drop(business_data.columns[[1,5,6]], axis=1) #dropped attributes,hours, is_open columns
yelp_df
```

```python
cols = yelp_df.columns.tolist()
cols
```

```python
col_list=pd.DataFrame({'col_number': range(11), 'col_name': cols})
col_list
```
