```python
import pandas as pd
```

```python
restaurant_df = pd.read_csv('Resources/DOHMH_New_York_City_Restaurant_Inspection_Results.csv')
restaurant_df.head()
```

```python
restaurant_df.info()
```

```python
## Drop unecessary columns
to_drop = ['PHONE', 'BUILDING','Community Board', 'Council District','Census Tract', 'BIN', 'BBL', 'NTA', 'CRITICAL FLAG', 'SCORE', 'GRADE DATE','RECORD DATE', 'INSPECTION TYPE']
new_restaurant_df = restaurant_df.drop(to_drop, axis=1)
new_restaurant_df.head()
```

```python
# Sum of missing values in each column
new_restaurant_df.isnull().sum()
```

```python
# Count the amount of rows that have duplication
new_restaurant_df.duplicated().sum()
```

```python
#Show the rows that have duplication
new_restaurant_df.loc[new_restaurant_df.duplicated(), :]
```

```python
# Drop null rows
final_restaurant_df = new_restaurant_df.dropna()
final_restaurant_df
```

```python
# After removing unecessary columns and nulll rows, the dataset is even.
final_restaurant_df.info()
```

```python
# Final df has no duplications
final_restaurant_df.duplicated().sum()
```

```python
# Final dataframe with the proposed columns dropped and removed NAs
final_restaurant_df.head()
```

```python
# Final dataframe with the proposed columns dropped, removed NAs and reformatted the ZIPCODE column to int32
final_restaurant_df.ZIPCODE = final_restaurant_df.ZIPCODE.astype('int')
final_restaurant_df.head()
```
